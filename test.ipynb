```python
import requests 
import os
from dotenv import load_dotenv
import pandas as pd

load_dotenv()
API_KEY = os.getenv('API_KEY')

url = "https://backend.simfin.com/api/v3/companies/prices/compact?ticker=MSFT"

headers = {
    "accept": "application/json",
    "Authorization": API_KEY
}

response = requests.get(url, headers=headers).json()
```

```python
json_data = response[0]
df = pd.DataFrame(json_data['data'], columns=json_data['columns'])
df.head()
```

```python
latest = df.tail(2)
latest
```

```python
latest['% change'] = latest['Last Closing Price'].pct_change()
latest
```

```python
latest['change'] = latest['Last Closing Price'].diff()
latest
```

```python
df_final = latest.tail(1)
df_final['% change'].iloc[0]
```
